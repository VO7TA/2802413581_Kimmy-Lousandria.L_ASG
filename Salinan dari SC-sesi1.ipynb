```python
def calculate(UAS, UTS, ASG):
    return (UAS * 0.5) + (UTS * 0.3) + (ASG * 0.2)

def grading(total):
    if 90 <= total <= 100:
        return "A"
    elif 85 <= total < 90:
        return "A-"
    elif 80 <= total < 85:
        return "B+"
    elif 75 <= total < 80:
        return "B"
    elif 70 <= total < 75:
        return "B-"
    elif 65 <= total < 70:
        return "C"
    elif 60 <= total < 65:
        return "D"
    elif 11 <= total < 60:
        return "E"
    else:
        return "F"

def scores():
    while True:
        try:
            UAS = float(input("Masukkan nilai UAS [0-100]: "))
            if 0 <= UAS <= 100:
                break
            else:
                print("Nilai harus dalam rentang 0 - 100. Coba lagi!")
        except ValueError:
            print("Input harus berupa angka. Coba lagi!")

    while True:
        try:
            UTS = float(input("Masukkan nilai UTS [0-100]: "))
            if 0 <= UTS <= 100:
                break
            else:
                print("Nilai harus dalam rentang 0 - 100. Coba lagi!")
        except ValueError:
            print("Input harus berupa angka. Coba lagi!")

    while True:
        try:
            ASG = float(input("Masukkan nilai ASIGMENT [0-100]: "))
            if 0 <= ASG <= 100:
                break
            else:
                print("Nilai harus dalam rentang 0 - 100. Coba lagi!")
        except ValueError:
            print("Input harus berupa angka. Coba lagi!")

    return UAS, UTS, ASG

def main():
    username = input("Masukkan username: ")
    while True:
        password = input("Masukkan password: ")
        if password == "Admin123":
          print(f"Login Berhasil, Selamat datang {username} ")
          break
        print("Password salah, coba lagi!")

    print("\n")
    nama_mahasiswa = input("Masukkan nama mahasiswa: ")
    UAS, UTS, ASG = scores()

    final_score = calculate(UAS, UTS, ASG)
    grade = grading(final_score)

    print("\n")
    print("============================================")
    print(f"Hasil perolehan nilai {nama_mahasiswa}")
    print(f"Final Score: {final_score:.2f}")
    print(f"Grade: {grade}")
    print("============================================")

if __name__ == "__main__":
    main()
```
